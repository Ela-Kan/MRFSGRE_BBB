```python
import os

# Load your data
data = []
labels = []
for f in os.listdir('DictionaryFISP'):
    if f.endswith('.npy'):
        array = np.load(os.path.join('DictionaryFISP', f))
        data.append(array)
        # Extract the second from last number from the file name
        number = float(f.split('_')[-2])
        labels.append(str(number))

# Create a new figure
plt.figure()

# Plot each array
for array, label in zip(data, labels):
    plt.plot(array/10000, label=label)

# Add a legend
legend = plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
legend.set_title("B1 Multiplication Factor Values")
plt.title('FISP Sequence Fingerprint With One Compartment (T1 = 700ms T2 = 85ms)')
plt.xlim([0,1000])

# Show the plot
plt.show()

```
